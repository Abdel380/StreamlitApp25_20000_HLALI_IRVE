```python
from utils.cleaning import clean_irve

df_clean = clean_irve("data/irve.csv")
df_clean.head()
df_clean.info()
```

```python
import matplotlib.pyplot as plt

# 1️⃣ Distribution des puissances
df_clean["puissance_kw"].plot(kind="hist", bins=50, figsize=(8,4))
plt.title("Distribution des puissances (kW)")
plt.xlabel("Puissance (kW)")
plt.ylabel("Nombre de bornes")
plt.show()
```

```python
# 2️⃣ Répartition par catégorie
df_clean["power_cat"].value_counts().plot(kind="bar", color="skyblue")
plt.title("Répartition des catégories de puissance")
plt.xlabel("Catégorie")
plt.ylabel("Nombre de bornes")
plt.show()
```

```python
# 3️⃣ Fiabilité estimée
df_clean["fiabilite_estimee"].value_counts().plot(kind="pie", autopct="%1.1f%%", figsize=(5,5))
plt.title("Niveau de fiabilité estimée des bornes")
plt.ylabel("")
plt.show()
```
